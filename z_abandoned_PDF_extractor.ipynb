```python
from bs4 import BeautifulSoup, SoupStrainer
import requests
import re
from os.path import basename, join
import datetime
from tabula import read_pdf

#example url
url = "http://www.rms.nsw.gov.au/projects/sydney-south/m5-east/outside-air-quality-monitoring/historical-data-"
target_class='panel-group'

year_start = 2000
now = datetime.datetime.now()
year_end = now.year

url_list = []

# build year list
for year in range(year_start, year_end + 1):
    data_url = url + str(year) + '.html'
    url_list.append(data_url)
```

```python
print(url_list)
```

```python
link_list = []

# loop through pages
for data_url in url_list:
    r = requests.get(data_url)
    soup = BeautifulSoup(r.content, "lxml")
    soup.find_all('div', class_=target_class)

    for a in soup.find_all('a', href=True):
        if "document" in a['href']:
            link_list.append("http://www.rms.nsw.gov.au"+a['href'])
```

```python
link_list
```

```python
# filter the list
link_list = [k for k in link_list if 'graph' not in k]
link_list = [k for k in link_list if 'portal' not in k]
link_list = [k for k in link_list if 'non' not in k]
link_list = [k for k in link_list if 'avocs' not in k]
link_list = [k for k in link_list if 'vent' not in k]
link_list = [k for k in link_list if 'fullreport' not in k]
link_list = [k for k in link_list if 'exceed' not in k]
```

```python
link_list
```

```python
doc_list = []

for url in link_list:
    response = requests.get(url)
    with open(join('./data-dump', basename(url)), 'wb') as f:
        f.write(response.content)

    doc_list.append(basename(url))
```
